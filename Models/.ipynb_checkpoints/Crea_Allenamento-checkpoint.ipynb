```python
df = pd.read_csv("./Data/Tel/Primo test/AccelerometerUncalibrated.csv")
```

```python
df_acc = df[["AccX", "AccY", "AccZ"]]
```

```python
df_gyro = df[["GyroX", "GyroY", "GyroZ"]]
```

```python
def high_pass_filter(data, alpha=0.9):
    # Inizializza il valore filtrato con il primo elemento di data
    filtered_data = [data[0]]

    # Usa una variabile temporanea per memorizzare il valore precedente del filtro
    previous_filtered = data[0]

    for i in range(1, len(data)):
        # Calcola il nuovo valore filtrato e aggiorna la variabile temporanea
        previous_filtered = alpha * (previous_filtered + data[i] - data[i - 1])
        filtered_data.append(previous_filtered)

    return filtered_data

def low_pass_filter(data, beta=0.1):
    # Inizializza il valore filtrato con il primo elemento di data
    filtered_data = [data[0]]

    # Usa una variabile temporanea per memorizzare il valore precedente del filtro
    previous_filtered = data[0]

    for i in range(1, len(data)):
        # Calcola il nuovo valore filtrato e aggiorna la variabile temporanea
        previous_filtered = beta * data[i] + (1 - beta) * previous_filtered
        filtered_data.append(previous_filtered)

    return filtered_data

```

```python
ALPHA = 0.95
BETA = 0.15

acc_high_filtered = {
    "AccX" : high_pass_filter(df["AccX"], ALPHA),
    "AccY" : high_pass_filter(df["AccY"], ALPHA),
    "AccZ" : high_pass_filter(df["AccZ"], ALPHA)            
                }

acc_high_filtered = pd.DataFrame(acc_high_filtered)

acc_filtered_final = {
    "AccX" : low_pass_filter(acc_high_filtered["AccX"], BETA),
    "AccY" : low_pass_filter(acc_high_filtered["AccY"], BETA),
    "AccZ" : low_pass_filter(acc_high_filtered["AccZ"], BETA)            
                }


acc_filtered_final = pd.DataFrame(acc_filtered_final)
```

```python
import matplotlib.pyplot as plt

# Grafico dei valori di accX, accX filtrato
plt.figure(figsize=(12, 6))
plt.plot(df['AccX'], label='AccX', color='blue', alpha=0.5)
plt.plot(acc_filtered_final['AccX'], label='AccX Filtered', color='red')
plt.title('Accelerazione X - Confronto Originale vs Filtrata')
plt.xlabel('Campioni')
plt.ylabel('Accelerazione')
plt.legend()
plt.show()
```

```python
# Calcolo vettorializzato della magnitudo per entrambi i dataframe
magnitude_not_filtered = np.sqrt(df["AccX"]**2 + df["AccY"]**2 + df["AccZ"]**2).to_list()
magnitude_filtered = np.sqrt(acc_filtered_final["AccX"]**2 + acc_filtered_final["AccY"]**2 + acc_filtered_final["AccZ"]**2).to_list()
```

```python
import matplotlib.pyplot as plt

# Grafico dei valori di accX, accX filtrato
plt.figure(figsize=(12, 6))
plt.plot(magnitude_not_filtered[0:1000], label='Magnitude not Filtered', color='blue', alpha=0.5)
plt.plot(magnitude_filtered[0:1000], label='Magnitude Filtered', color='red')
plt.title('Magnitudine - Confronto Originale vs Filtrata')
plt.xlabel('Campioni')
plt.ylabel('Magnitudine')
plt.legend()
plt.show()
```

```python
# Soglia universale per movimento minimo
soglia_universale = 0.70  # adatta tra 0.1 e 0.3 m/s secondo la tua sensibilit

df_acc['is_moving'] = list(map(lambda x: x > soglia_universale, magnitude_filtered))
```

```python
df_acc.to_csv("./Data/File_Acc_Tel_v1.csv", index=False)
```
